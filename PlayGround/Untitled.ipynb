```python
import pandas as pd
import sklearn
from sklearn.linear_model import *
from sklearn.metrics import *
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
X, y = sklearn.datasets.make_regression(1000,2,noise=0)
```

```python
y.shape
```

```python
X
```

```python
figure, (ax1,ax2) = plt.subplots(1,2,figsize=(15,5))

ax1.scatter(X[:,0],y)
ax2.scatter(X[:,1],y)

plt.show()
```

```python
linmodel = LinearRegression()

linmodel.fit(X,y)

y_pred = linmodel.predict(X)

#plt.scatter(X,y)

#plt.plot(X,y_pred, color="red")

((y - y_pred)**2).mean()

```

```python
mean_squared_error(y,y_pred)
```
