```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy.linalg as LA
import numpy as np
u = np.array([2, 5])
v = np.array([3, 1])
def plot_vector2d(vector2d, origin=[0, 0], **options):
    return plt.arrow(origin[0], origin[1], vector2d[0], vector2d[1],
              head_width=0.2, head_length=0.3, length_includes_head=True,
              **options)

```

```python
x_coords, y_coords = zip(u, v)
plt.scatter(x_coords, y_coords, color=["r","b"])
plt.axis([0, 9, 0, 6])
plt.grid()
plt.show()
```

```python
radius = LA.norm(v)

plt.gca().add_artist(plt.Circle((0,0), radius,alpha=0.4, color="#DDDDDD"))
plot_vector2d(v, color="blue")
plt.axis([0, 8.7, 0, 6])
plt.grid()
plt.show()
```

```python
print(v+u)
plot_vector2d(v, color="blue")
plot_vector2d(u, color="red")
plot_vector2d(u+v, color="purple")
plt.axis([0, 10, 0, 10])
plt.grid()
plt.show()
```

```python
print(v)
print(u)
print(v*u)
plot_vector2d(v, color="blue")
plot_vector2d(u, color="red")
plot_vector2d(u*v, color="purple")
plt.axis([0, 10, 0, 10])
plt.grid()
plt.show()
```

```python
def vector_angle(u, v):
    cos_theta = u.dot(v) / (LA.norm(u) * LA.norm(v))
    return np.arccos(np.clip(cos_theta, -1, 1))

theta = vector_angle(u, v)
print("Angle =", theta, "radians")
print("      =", theta * 180 / np.pi, "degrees")
```

```python

v_normalized = v / LA.norm(v)
print(v_normalized)
u_normalized = u / LA.norm(u)
print(u_normalized)

projbas = u.dot(v)/(LA.norm(u)**2)*u

proj = v.dot(u_normalized) * u_normalized
print(projbas)
print(proj)



print(f'length projected line: {LA.norm(proj)}')
print(f'length u line: {LA.norm(u)}')
print(f'product lengths: {LA.norm(proj)*LA.norm(u)}')
print(f'dotproduct: {v.dot(u)}')
print(f'dotproduct: {v.dot(u)/LA.norm(u)}')

plot_vector2d(v, color="blue")
plot_vector2d(u, color="red")
#plot_vector2d(v_normalized, color="purple") # normalized v
#plot_vector2d(u_normalized, color="orange")

plot_vector2d(proj, color="k", linestyle=":")

plt.plot([proj[0], v[0]], [proj[1], v[1]], "b:")

plt.axis([0, 8, 0, 5.5])
plt.grid()
plt.show()
```
